```python
# library imports
import geopandas as gpd
from quetzal.model import stepmodel
# Notebook setting to show plots under the cells
%matplotlib inline
```

```python
# Load zoning data file that came with the Paris example
zones = gpd.read_file('../inputs/zones/zones.geojson')
# How does the DataFrame look like?
zones.sample()
```

```python
# We can plot the geometry column
zones.plot()
```

```python
# Now create a quetzal model instance with Coordinate Reference System 4326
sm = stepmodel.StepModel(epsg=4326, coordinates_unit='degree')
# See what attributes this model has
sm.__dict__.keys()
```

```python
# But all of these tables are empty
sm.zones
```

```python
# Save the Paris zones as model zones
sm.zones = zones
```

```python
# Save the model to geojson files
sm.to_json('../../model/zones')
```
